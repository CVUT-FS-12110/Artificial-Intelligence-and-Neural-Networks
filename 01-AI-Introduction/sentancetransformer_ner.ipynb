# Transformers for text processing

## Text vectorization

```python
from sentence_transformers import SentenceTransformer


# Load model
model = SentenceTransformer('all-mpnet-base-v2')

# Sentences for vectorization
sentences = [
    "This is the first sentence.",
    "Here is another sentence.",
    "Sentence transformers are very useful.",
    "This is different sentence"
]

# Vectorization
sentence_embeddings = model.encode(sentences)

# Output
for sentence, embedding in zip(sentences, sentence_embeddings):
    print(f"Sentence: {sentence}")
    print(f"Embedding: {embedding[:5]}...")
    print()
```

## Cosine similiarity of vectorizet text

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity([sentence_embeddings[1]], sentence_embeddings)
```

```python
cosine_similarity([sentence_embeddings[1]], sentence_embeddings)
```

## Named Entity Recognition - NER

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline

# Load the model
model_name = "dslim/bert-base-NER"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

# Set NER pipeline
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)


# Text processing
entities = ner_pipeline(history)

# List of named entities
print("Pojmenované entity:")
for entity in entities:
    print(f"Text: {entity['word']}, Label: {entity['entity']}")
```
